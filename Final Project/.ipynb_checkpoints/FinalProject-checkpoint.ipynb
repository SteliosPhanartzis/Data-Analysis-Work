# The Dataset
Reference to [similar dataset](http://cseweb.ucsd.edu/classes/sp15/cse190-c/reports/sp15/048.pdf) for additonal information on columns in our df. Definition of headings given below.

* **age:** the age of an individual

* **workclass:** a general term to represent the employment status of an individual

* **fnlwgt:** final weight. In other words, this is the number of people the census believes the entry represents.

* **education:** the highest level of education achieved by an individual.

* **education­num:** the highest level of education achieved in numerical form.

* **marital­status:** marital status of an individual. Married­civ­spouse corresponds to a
civilian spouse while Married­AF­spouse is a spouse in the Armed Forces.

* **occupation:** the general type of occupation of an individual

* **relationship:** represents what this individual is relative to others. For example an
individual could be a Husband. Each entry only has one relationship attribute and is
somewhat redundant with marital status. We might not make use of this attribute at all

* **race:** Descriptions of an individual’s race

* **sex:** the biological sex of the individual

* **capital­gain:** capital gains for an individual

* **capital­loss:** capital loss for an individual

* **hours­per­week:** the hours an individual has reported to work per week

* **native­country:** country of origin for an individual

* **salary:** whether or not an individual makes more than $80,000 annually.

# Data Normalization

```python
# Copy df to df2, normalize columns if all values are numeric
df2 = df.copy()
for feature_name in df.columns:
    max_value = df[feature_name].max()
    min_value = df[feature_name].min()
    df2[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
```

## Check logistic regression model accuracy with all features

```python
# Split data as 70% train / 30% test
X_train, X_test, y_train, y_test = train_test_split(df2.drop('salary',axis=1), df2['salary'], test_size=0.30, random_state=101)
model = LogisticRegression()
model.fit(X_train,y_train)
predictions = model.predict(X_test)
print(classification_report(y_test,predictions))
print(accuracy_score(y_test, predictions))
```

# Train/Test Split

```python
# Split data as 70% train / 30% test
X_train, X_test, y_train, y_test = train_test_split(df2.drop('salary',axis=1), df2['salary'], test_size=0.30, random_state=101)
```

# Testing Various Models

```python
model = LogisticRegression()
model.fit(X_train,y_train)
predictions = model.predict(X_test)
print(classification_report(y_test,predictions))
print(accuracy_score(y_test, predictions))
```

```python
# Gamma of 10, C of 10
model = svm.SVC(kernel='rbf', gamma=10, C=10)
# Retrain model
model.fit(X_train,y_train)
#Save model results
predictions = model.predict(X_test)
#Evaluate and print accuracy score
print(classification_report(y_test,predictions))
print(accuracy_score(y_test, predictions))
```

```python
# max_depth of 10, criterion = gini
model = DecisionTreeClassifier(max_depth=10, criterion='entropy')
# Train model
model.fit(X_train,y_train)
#Save model results
predictions = model.predict(X_test)
#Evaluate and print accuracy score
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
print(accuracy_score(y_test, predictions))
```

```python
# max_depth of 10, criterion = gini
model = DecisionTreeClassifier(max_depth=14, criterion='gini')
# Train model
model.fit(X_train,y_train)
#Save model results
predictions = model.predict(X_test)
#Evaluate and print accuracy score
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
print(accuracy_score(y_test, predictions))
```

```python
# Random Forest Classifier
rfc = RandomForestClassifier(n_estimators=80)
rfc.fit(X_train, y_train)
rfc_pred = rfc.predict(X_test)
print(classification_report(y_test, rfc_pred))
print(accuracy_score(y_test, rfc_pred))
```

# K-Fold Cross Validation

```python
from sklearn.model_selection import cross_val_score
df2 = df.copy()
x = df2.drop(columns=["salary"])
y = df2["salary"]
results = {}
# Split data as 70% train / 30% test
X_train, X_test, y_train, y_test = train_test_split(df2.drop('salary',axis=1), df2['salary'], test_size=0.30, random_state= random.randint(1,1000))
for i in range(1,11):
    # Random Forest Classifier
    model = RandomForestClassifier(n_estimators=(i*10))
    # Save model results, perform 10-fold cross validation 
    scores = cross_val_score(estimator=model, X=x, y=y, cv=10, n_jobs=4)
    results[i*10] = scores.mean()
```
