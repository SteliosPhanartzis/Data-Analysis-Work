# Question 10:  Use that variable to build a one-variable Linear Regression model of the house price. Make sure you split the data between Training and Testing set first, Save 20% as your testing data

```python
# Show your work here
# Get Columns
income = housing['Income'].values.reshape(-1,1)
price = housing['Price'].values.reshape(-1,1)
# Split data and confirm
income_trn, income_tst, price_trn, price_tst = train_test_split(income, price, test_size=0.2, random_state=0)
print("income_trn size: %s income_tst size: %s" % (income_trn.shape, income_tst.shape))
print("price_trn size: %s price_tst size: %s" % (price_trn.shape, price_tst.shape))
# Create models
model1 = LinearRegression()
model2 = LinearRegression()
# Train models
model1.fit(income, price)
model2.fit(income_trn,price_trn)
print("intercept: %s coef: %s" % (model2.intercept_, model2.coef_))
# Predict output on test
price_fullPred = model1.predict(income)
price_splitpRed = model2.predict(income_tst)
```

```python
# Show your R-square and RMSE
# Train
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(price, price_fullPred)))
print('R-squared:', metrics.r2_score(price, price_fullPred))
# Test
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(price_tst, price_splitPred)))
print('R-squared:', metrics.r2_score(price_tst, price_splitPred))
```

```python
# Plot data for visual context
plt.scatter(income_tst, price_tst,  color='gray')
plt.plot(income_tst, price_splitPred, color='red', linewidth=2)
plt.show()
```

# Question 11: Now do a 10-fold Cross Validation of your model. Does your model pass cross validation?

The output of the model given random data remains fairly constant and within range, so it passes cross validation.

```python
# Show your work here
rsquare = []
for i in range(10):
    price_train, price_test, income_train, income_test = train_test_split(price, income, test_size=0.2, random_state = random.randint(1,1000))
    model = LinearRegression()
    model.fit(price_train, income_train)
    income_pred = model.predict(price_test)
    rsquare.append(metrics.r2_score(income_test, income_pred))
    
rdf = pd.DataFrame( {'trial': range(10), 'Rsquare': rsquare})
rdf
```

```python
# Plot bar graph
rdf.plot.bar(x='trial', y='Rsquare');
```

# Question 12: Now pick ONE more variable in addition to the variable you choose from Question 9 and build a two-variables Linear Regression model of the house price. Make sure you split the data into training and testing set first

```python
# Show your work here
# Get Columns
twoColVar = housing[['Income','Population']].values.reshape(-1,2)
price = housing['Price'].values.reshape(-1,1)
# Split data and confirm
twoColVar_trn, twoColVar_tst, price_trn, price_tst = train_test_split(twoColVar, price, test_size=0.2, random_state=0)
print("twoColVar_trn size: %s twoColVar_tst size: %s" % (twoColVar_trn.shape, twoColVar_tst.shape))
print("price_trn size: %s price_tst size: %s" % (price_trn.shape, price_tst.shape))
# Create models
model1 = LinearRegression()
model2 = LinearRegression()
# Train models
model1.fit(twoColVar, price)
model2.fit(twoColVar_trn, price_trn)
print("intercept: %s coef: %s" % (model2.intercept_, model2.coef_))
# Predict output on test
price_fullPred = model1.predict(twoColVar)
price_splitPred = model2.predict(twoColVar_tst)
```

```python
# Show your R-square and RMSE
# Train
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(price, price_fullPred)))
print('R-squared:', metrics.r2_score(price, price_fullPred))
# Test
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(price_tst, price_splitPred)))
print('R-squared:', metrics.r2_score(price_tst, price_splitPred))
```

```python
# Plot data for visual context
plt.plot(twoColVar_tst, price_splitPred, color='red', linewidth=2)
plt.show()
```

# Question 13: Do a 10-fold Cross Validation of your two-variable model. Does your model pass cross validation?

The model's rsquare is fairly lower on randomized input, suggesting that the model was overfitting on the training data.

```python
# Show your work here
rsquare = []
for i in range(10):
    price_train, price_test, twoColVar_train, twoColVar_test = train_test_split(price, twoColVar, test_size=0.2, random_state = random.randint(1,1000))
    model = LinearRegression()
    model.fit(price_train, twoColVar_train)
    twoColVar_pred = model.predict(price_test)
    rsquare.append(metrics.r2_score(twoColVar_test, twoColVar_pred))
    
rdf = pd.DataFrame( {'trial': range(10), 'Rsquare': rsquare})
rdf
```

```python
# Plot bar graph
rdf.plot.bar(x='trial', y='Rsquare');
```
